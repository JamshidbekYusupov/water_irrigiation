```python
import pandas as pd
import numpy as np 
import shap
import joblib
```

```python
# Loading data
test_path = r'C:\Irrigation_Water_Requirement\Data\engineered_Data\X_test.csv'
y_path = r'C:\Irrigation_Water_Requirement\Data\Preprocesssed_Data\missed_data\y_test.csv'

X_test = pd.read_csv(test_path)
y_test = pd.read_csv(y_path)
```

```python
X_test['Previous_Irrigation_mm'] = X_test['Previous_Irrigation_mm'].astype('object')
X_test['Wind_Speed_kmh'] = X_test['Wind_Speed_kmh'].astype('object')
X_test['Field_Area_hectare'] = X_test['Field_Area_hectare'].astype('object')
X_test['Field_Area_hectare'] = X_test['Field_Area_hectare'].astype('object')
```

```python
# Model loading 

model = r'C:\Irrigation_Water_Requirement\Models\all_models\DT.joblib'

model = joblib.load(model)
```

```python
model
```

```python
X_test_preprocessed = model.named_steps['preprocessor'].transform(X_test)
```

```python
y_pred = model.predict(X_test)
```

```python
explainer = shap.TreeExplainer(model.named_steps['model'])
shap_values = explainer.shap_values(X_test_preprocessed)
```

```python
shap.waterfall_plot(shap.Explanation(
    values=shap_values[0][0],
    base_values=explainer.expected_value[0],
    data=X_test_preprocessed[0],
    feature_names=X_test.columns
))
```

```python
# --- Summary plot ---
shap.summary_plot(shap_values, X_test_preprocessed, feature_names=X_test.columns)
```
